```python
df.sale_price=np.where(df.sale_price==22325.0,325.0,df.sale_price)
```

```python
df[df.sale_price>10000]
```

```python
#Checking Null Values in Sale_price Columns

df[df.sale_price.isnull()]
```

```python
#Fill null values in sale_price columns from Market_price columns

df.sale_price = df.sale_price.fillna(df.market_price)
```

```python
#Checking Null Values now.

df.iloc[1718:1724]
```

```python
# Final check null values in sale_price columns
df.sale_price.isnull().sum()
```

```python
# Handle missing values in 'product' and 'brand'
df['product'].fillna('Unknown Product',inplace=True)
df['brand'].fillna('Unknown Brand',inplace=True)

# Handle missing values in 'rating' by imputing with the median
df['rating'].fillna(df['rating'].median(),inplace=True)

# Handle missing values in 'description' by replacing with an empty string
df['description'].fillna('',inplace=True)
```

```python
#Checking Null Values now
df.isnull().sum()
```

### **Pie Chart for Product Categories:**

### **Category wise Discount**:

```python
# Category wise Discount
MRP= df.groupby('category')['market_price'].sum()
SRP =df.groupby('category')['sale_price'].sum()
(MRP-SRP)/MRP * 100
```

```python
MRP = pd.DataFrame(MRP).reset_index()
SRP = pd.DataFrame(SRP).reset_index()
Dis = pd.DataFrame({'category':MRP.category, 'Original Price':MRP.market_price,
                    'Sale Price':SRP.sale_price, 'Discount Price':MRP.market_price-SRP.sale_price,
                    'Disount Percent':(MRP.market_price-SRP.sale_price)/MRP.market_price*100
})
```

```python
Dis
```

```python
## Bar plot to show category-wise Discount
a =sns.barplot(y=Dis.category,x=Dis['Disount Percent'])
a.bar_label(a.containers[0])
plt.title('Category wise Discount')
```

### **Correlation Matrix for Numerical Variables:**

```python
# Select only numeric columns for correlation calculation
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the correlation matrix for numeric variables
correlation_matrix = df[numeric_columns].corr()

# Heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Numeric Variables')
plt.show()
```
